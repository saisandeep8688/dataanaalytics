
Exploratory Data Analysis (EDA)

```python
# prompt: Plot the distribution of the following discrete features:
# # o	Quantity
# # o	Gender
# # o	Transaction ID
# # for above dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data Loading and Inspection (as provided in the prompt)
try:
    df = pd.read_csv('/content/retail_sales_dataset.csv')
except FileNotFoundError:
    print("Error: '/content/retail_sales_dataset.csv' not found. Please upload the file or provide the correct path.")
    df = None

if df is not None:
    print(df.isnull().sum())
    # Check for duplicate rows
    print(df.duplicated().sum())
    # Drop duplicate rows
    df.drop_duplicates(inplace=True)
    # Display cleaned data information
    print("\nDataframe info after cleaning:")
    print(df.info())
    # Display descriptive statistics after cleaning
    print("\nDescriptive statistics after cleaning:")
    print(df.describe())

    # Exploratory Data Analysis (EDA)
    # Plotting distributions for discrete features
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    sns.countplot(x='Quantity', data=df)
    plt.title('Distribution of Quantity')

    plt.subplot(1, 3, 2)
    sns.countplot(x='Gender', data=df)
    plt.title('Distribution of Gender')

    plt.subplot(1, 3, 3)
    sns.countplot(x='Transaction ID', data=df)
    plt.title('Distribution of Transaction ID')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

    plt.tight_layout()
    plt.show()
```


Correlation Analysis

```python
# prompt: correlation matrix for numerical columns on above data set

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ... (Your existing code)

# Correlation Analysis
if df is not None:
    # Select numerical columns for correlation analysis
    numerical_cols = df.select_dtypes(include=['number'])

    # Calculate the correlation matrix
    correlation_matrix = numerical_cols.corr()

    # Visualize the correlation matrix using a heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Numerical Features')
    plt.show()

```


DataFrame Operations

```python
#  Use loc to extract sales data for a specific date or range of dates.
if df is not None and 'Date' in df.columns and 'Sales' in df.columns:
    # Example: Extract sales data for 2024-01-15
    specific_date_sales = df.loc[df['Date'] == '2024-01-15', 'Sales']
    print("\nSales on a specific date:")
    print(specific_date_sales)

    # Example: Extract sales data between 2024-01-01 and 2024-01-31
    start_date = '2024-01-01'
    end_date = '2024-01-31'
    date_range_sales = df.loc[(df['Date'] >= start_date) & (df['Date'] <= end_date), 'Sales']
    print("\nSales within a date range:")
    print(date_range_sales)

# Use iloc to extract the first 10 rows of the dataset.
if df is not None:
    first_10_rows = df.iloc[:10]
    print("\nFirst 10 rows:")
    print(first_10_rows)

#  Drop a column (e.g., Region) using drop(axis=1).
if df is not None and 'Region' in df.columns:
    df = df.drop('Region', axis=1)
    print("\nDataFrame after dropping the 'Region' column:")
    print(df.head())  # Display the first few rows to verify

# 4. Drop the first two rows using drop(axis=0).
if df is not None:
    df = df.drop(df.index[:2]) # Drop first two rows
    print("\nDataFrame after dropping the first two rows:")
    print(df.head()) # Display first few rows to verify
```
